```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color, filters, feature, exposure
from skimage.transform import resize
from skimage.measure import regionprops
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, MaxPooling2D, Lambda, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import confusion_matrix, classification_report, f1_score
from sklearn.utils import class_weight
```

```python
# Paths
genuine_image_paths = "C:/Users/intra/signatures_main/dataset1/real/"
forged_image_paths = "C:/Users/intra/signatures_main/dataset1/forge/"
```

```python
# Image preprocessing
def preprocess_image(img):
    if len(img.shape) == 3:
        img = color.rgb2gray(img)
    img = resize(img, (128, 128), anti_aliasing=True)  # Increased resolution
    img = exposure.equalize_adapthist(img)  # Contrast enhancement
    img = img.astype('float32')
    return img

def extract_features(img):
    if len(img.shape) != 2:
        raise ValueError("Input image must be 2D (grayscale)")
    
    # Enhanced feature extraction
    thresh = filters.threshold_otsu(img)
    binary = img > thresh
    
    props = regionprops(binary.astype(int))[0]
    
    # Adding more features
    hog_features = feature.hog(img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(3, 3))
    lbp = feature.local_binary_pattern(img, P=8, R=1)
    
    features = [
        props.area,
        props.eccentricity,
        props.extent,
        props.perimeter,
        props.solidity,
        props.euler_number,
        props.equivalent_diameter,
        np.mean(hog_features),
        np.std(hog_features),
        np.mean(lbp),
        np.std(lbp)
    ]
    
    return np.array(features)
```

```python
# Load and preprocess images
def load_and_preprocess_images(path, limit):
    images = []
    for filename in os.listdir(path)[:limit]:
        if filename.endswith((".png", ".jpg", ".jpeg")):
            img = io.imread(os.path.join(path, filename))
            img = preprocess_image(img)
            images.append(img)
    return np.array(images)
```

```python
# Load images (increased to 100 genuine and 100 forged for more data)
genuine_images = load_and_preprocess_images(genuine_image_paths, 100)
forged_images = load_and_preprocess_images(forged_image_paths, 100)
```

```python
# Extract features
genuine_features = np.array([extract_features(img) for img in genuine_images])
forged_features = np.array([extract_features(img) for img in forged_images])
```

```python
# Save features
np.save('genuine_features.npy', genuine_features)
np.save('forged_features.npy', forged_features)
```

```python
# Prepare data for Siamese Network
def create_pairs(x, y):
    pairs = []
    labels = []
    n = min(len(x), len(y))
    for i in range(n):
        # Positive pairs
        for j in range(i+1, n):
            pairs += [[x[i], x[j]]]
            labels += [1]
        # Negative pairs
        pairs += [[x[i], y[i]]]
        labels += [0]
    return np.array(pairs), np.array(labels)

pairs, labels = create_pairs(genuine_images, forged_images)

def create_siamese_model(input_shape):
    input = Input(input_shape)
    x = Conv2D(64, (3, 3), activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(input)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    
    x = Conv2D(128, (3, 3), activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    
    x = Conv2D(256, (3, 3), activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.3)(x)
    
    encoded = Model(inputs=input, outputs=x)
    
    input_a = Input(input_shape)
    input_b = Input(input_shape)
    
    processed_a = encoded(input_a)
    processed_b = encoded(input_b)
    
    distance = Lambda(lambda x: tf.math.abs(x[0] - x[1]))([processed_a, processed_b])
    
    output = Dense(1, activation='sigmoid')(distance)
    
    siamese_model = Model(inputs=[input_a, input_b], outputs=output)
    
    return siamese_model
```

```python
# Split data
train_pairs, test_pairs, train_labels, test_labels = train_test_split(pairs, labels, test_size=0.2, random_state=42)
```

```python
# Enhanced data augmentation
data_gen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.7, 1.3],
    fill_mode='nearest'
)
```

```python
# Create and compile the model
input_shape = (128, 128, 1)  # Updated for the new image size
model = create_siamese_model(input_shape)
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0001), metrics=['accuracy'])
```

```python
# Callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
model_checkpoint = ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_accuracy')
```

```python
# Reshape the input data
train_pairs = train_pairs.reshape(train_pairs.shape[0], 2, 128, 128, 1)
test_pairs = test_pairs.reshape(test_pairs.shape[0], 2, 128, 128, 1)
```

```python
# Calculate class weights
class_weights = class_weight.compute_class_weight(class_weight='balanced', 
                                                  classes=np.unique(train_labels), 
                                                  y=train_labels)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}
```

```python
# Train the model
history = model.fit(
    [train_pairs[:, 0], train_pairs[:, 1]], train_labels,
    validation_data=([test_pairs[:, 0], test_pairs[:, 1]], test_labels),
    batch_size=32,
    epochs=50,
    callbacks=[early_stopping, reduce_lr, model_checkpoint],
    class_weight=class_weight_dict
)
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate([test_pairs[:, 0], test_pairs[:, 1]], test_labels)
print(f"Test accuracy: {test_accuracy}")
```

```python
# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Find best threshold
def find_best_threshold(model, test_pairs, test_labels):
    predictions = model.predict([test_pairs[:, 0], test_pairs[:, 1]])
    thresholds = np.arange(0.1, 1.0, 0.01)
    best_threshold = 0.5
    best_f1 = 0
    
    for threshold in thresholds:
        y_pred = (predictions > threshold).astype(int)
        f1 = f1_score(test_labels, y_pred)
        if f1 > best_f1:
            best_f1 = f1
            best_threshold = threshold
    
    return best_threshold

best_threshold = find_best_threshold(model, test_pairs, test_labels)
print(f"Best threshold: {best_threshold}")
```

```python
# Use the best threshold for final evaluation
y_pred = (model.predict([test_pairs[:, 0], test_pairs[:, 1]]) > best_threshold).astype(int).flatten()

print("Confusion Matrix:")
print(confusion_matrix(test_labels, y_pred))

print("\nClassification Report:")
print(classification_report(test_labels, y_pred))
```

```python
# Function to test an unknown signature
def test_unknown_signature(model, unknown_signature_path, known_signature_path, threshold):
    # Load and preprocess the unknown signature
    unknown_img = io.imread(unknown_signature_path)
    unknown_img = preprocess_image(unknown_img)
    unknown_img = unknown_img.reshape(1, 128, 128, 1)  
    
    # Load and preprocess a known signature (for comparison)
    known_img = io.imread(known_signature_path)
    known_img = preprocess_image(known_img)
    known_img = known_img.reshape(1, 128, 128, 1)  
    
    # Make predictions
    prediction = model.predict([unknown_img, known_img])
    
    # if the prediction is greater than the threshold
    result = "Genuine" if prediction > threshold else "Forged"
    
    return result, prediction

# Example usage
unknown_signature_path = "path_to_unknown_sign"  # Path to the unknown signature
known_signature_path = "C:/Users/intra/signatures_main/dataset1/real/01200012.png"  

# Test the unknown signature
result, prediction = test_unknown_signature(model, unknown_signature_path, known_signature_path, best_threshold)
print(f"The unknown signature is: {result} (Prediction probability: {prediction[0][0]:.4f})")
```
